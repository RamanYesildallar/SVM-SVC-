```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Invistico_Airline.csv")
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
encoder=LabelEncoder()
k=encoder.fit_transform(df["satisfaction"])
k=pd.Series(k)
```

```python
newdf=df.iloc[:,9:21]
```

```python
newdf["satisfaction"]=k
```

```python
newdf
```

```python
shortdf=newdf.iloc[0:129880:7]
```

```python
shortdf
```

```python
X=shortdf.iloc[:, :-1].values
y=shortdf.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
x_test=sc.transform(X_test)
```

```python
from sklearn.svm import SVC
clasifier=SVC(kernel="rbf",random_state=0)
clasifier.fit(X_train,y_train)
```

```python
y_pred=clasifier.predict(sc.transform(X_test))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```
